```python
import imageio.v3 as iio
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
```

```python
img = iio.imread(uri="assets/solo_levelling.png")
img
```

```python
# this takes the first channel the Red channel 
_newimg_red = img[:,:,0]
_newimg_red = np.where(_newimg_red < 50, 0, _newimg_red)
# remove too high
_newimg_red = np.where(_newimg_red > 200, 0, _newimg_red)

_newimg_green = img[:,:,1]
_newimg_green = np.where(_newimg_green < 55, 0, _newimg_green)

_newimg_blue = img[:,:,2]
_newimg_blue = np.where(_newimg_blue < 50, 0, _newimg_blue)
```
